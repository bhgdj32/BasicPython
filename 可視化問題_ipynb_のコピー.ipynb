## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](https://github.com/shinonome-inc/PythonTutorialForDSCourse/blob/main/exercise/images/plot_1.png?raw=1)

```python
df = pd.read_csv("hist_data.csv")
plt.title("aデータとbデータの温度に関する度数分布")
plt.xlabel("温度"),plt.ylabel("度数")
plt.xlim(0,55),plt.ylim(0,30)
plt.grid("True")
plt.hist(df["aデータ"], bins= 30 ,label="aデータ",alpha=0.5, color= "orange")
plt.hist(df['bデータ'], bins=30, label='bデータ', alpha=0.3, color='cyan')
plt.show
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](https://github.com/shinonome-inc/PythonTutorialForDSCourse/blob/main/exercise/images/plot_2.png?raw=1)

```python
df2 = pd.read_csv("scatter_data.csv")
plt.title("クラス０～2のlengthとeidthに関する散布図")
plt.xlabel("length(cm)"), plt.ylabel("width(cm)")
plt.xlim(4.0,8.0),plt.ylim(1.7,4.5)
plt.grid("True")
class0 = df2[df2['class'] == 0]
class1 = df2[df2["class"] == 1]
class2 = df2[df2["class"] == 2]
plt.scatter(class0['length (cm)'], class0['width (cm)'], label='class0', color='r')
plt.scatter(class1["length (cm)"], class1["width (cm)"],label = "class1",color ="b")
plt.scatter(class2["length (cm)"], class2["width (cm)"], label = "class2",color = "g")
plt.show
```

使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_2](https://github.com/shinonome-inc/PythonTutorialForDSCourse/blob/main/exercise/images/plot_3_2.png?raw=1)

```python
df3 = pd.read_csv("line_data.csv")
plt.title("期間ごとの件数の推移")
plt.xlabel("期間"),plt.ylabel("件数")
plt.xticks(rotation=30), plt.ylim(100,150)
plt.grid(True)
plt.plot(df3["期間"],df3["件数"],marker='o')
plt.show
```

```python
df4 = pd.read_csv("line_data.csv")
plt.title("期間ごとの費用の推移")
plt.xlabel("期間"),plt.ylabel("費用")
plt.xticks(rotation = 30),plt.ylim(0,600000)
plt.grid(True)
plt.bar(df4["期間"],df4["費用"], width =0.8 )
plt.show
```
